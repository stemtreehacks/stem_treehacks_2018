```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import arrow
from dateutil import parser
import math
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
from sklearn import preprocessing
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from sklearn.preprocessing import MinMaxScaler
data=pd.read_csv('data.csv', sep=',')
data.columns = ['timestamp', 'avg_power']
data['timestamp'] = data['timestamp'].apply(arrow.get)
```

# Preliminary Data Investigation

```python
# Utility functions to query the data set and to plot and print results.

def get_points(time1, time2, groupbydayofweek=False, groupbyday=False, groupbytime=False):
    if groupbydayofweek:
        dataframe = data.copy()
        dataframe['day'] = dataframe['timestamp'].apply(lambda x: int(x.format('d')))
        indices = dataframe
    
    elif groupbyday:
        startindex = arrow.get(time1)
        stopindex = arrow.get(time2)
    
        indices = data.where((data['timestamp'] > startindex) & (data['timestamp'] < stopindex))
        indices = indices[np.isfinite(indices['avg_power'])]
        
    elif groupbytime:
        dataframe = data.copy()
        dataframe['hour'] = dataframe['timestamp'].apply(lambda x: int(x.format('H')))
        starttime = int(time1[:2])
        endtime = int(time2[:2])
        indices = data.where((dataframe['hour'] >= starttime) & (dataframe['hour'] < endtime))
        indices = indices[np.isfinite(indices['avg_power'])]
        
    return indices

def plot_points(time1, time2, groupbydayofweek=False, groupbyday=False, groupbytime=False):
    points = get_points(time1, time2, groupbydayofweek, groupbyday, groupbytime)
    if groupbydayofweek:
        points['timestamp'] = points['timestamp'].apply(lambda x: x.format('d'))
        points = points.groupby('timestamp').agg('mean')
        points['timestamp'] = points.index
        points = points.sort_values('timestamp')
        plt.plot(points['timestamp'], points['avg_power'])
        print(points)
        
    elif groupbyday:
        points['timestamp'] = points['timestamp'].apply(lambda x: x.format('DDDD'))
        points = points.groupby('timestamp').agg('mean')
        points['timestamp'] = points.index
        plt.plot(range(len(points)), points['avg_power'], label=time1 + " - " + time2)
        print(time1, 'to', time2, ':', np.nanmean(points['avg_power']))

    elif groupbytime:
        print(time1, ' to ', time2)
        points['month'] = points['timestamp'].apply(lambda x: int(x.format('M')))
        points['timestamp'] = points['timestamp'].apply(lambda x: x.format('M H:mm'))
        points = points.groupby('timestamp').agg('mean')
        points['timestamp'] = points.index
        for month in range(1, 13):
            p = points.where(points['month'] == month)
            plt.plot(range(len(p)), p['avg_power'], label=month)
            print(month, ': ', np.nanmean(p['avg_power']))

    plt.legend()
    return points['avg_power']
```

```python
# Average energy usage for the 4 seasons.
seasons = [
    ["2017-01-01 00:00", "2017-03-21 00:00"],
    ["2017-03-21 00:00", "2017-06-21 00:00"],
    ["2017-06-01 00:00", "2017-09-21 00:00"],
    ["2017-09-01 00:00", "2018-12-21 00:00"]
]

for s in seasons:
    plot_points(s[0], s[1], groupbyday=True)
plt.title('Energy usage during each season')
plt.show()
```

```python
# Average energy usage during each time of day, separated by month.
timeofday = [
    ["00:00", "06:00"],
    ["06:00", "12:00"],
    ["12:00", "18:00"],
    ["18:00", "24:00"]
]

for t in timeofday:
    np.mean(plot_points(t[0], t[1], groupbytime=True))
    plt.title('Energy usage during: ' + t[0] + ' to ' + t[1])
    plt.show()
```

```python
# Average energy usage for each day of the week across the entire year.
# Below, 1=Monday, 2=Tuesday, ..., 7=Sunday.
np.mean(plot_points(None, None, groupbydayofweek=True))
plt.title('Energy usage during each day of the week')
plt.show()
```

# Peak Smoothing to lower Demand Charges

## Calculating Demand Charges

```python
def compute_demand_charge(dataframe, energy_column='avg_power', charge_per_kw=35):
    dataframe = dataframe.copy()
    result = pd.DataFrame()
    dataframe['timestamp'] = dataframe['timestamp'].apply(lambda x: x.format('M'))
    dataframe = dataframe.groupby('timestamp').agg('max')
    dataframe['Month'] = dataframe.index
    result['Month'] = dataframe['Month'].apply(lambda x: int(x))
    result = result.sort_values('Month')
    result['Demand Charge'] = dataframe[energy_column]*charge_per_kw
    result.index.name = None
    
    return result
```

```python
# Demand costs for every month without peak smoothing
compute_demand_charge(data)
```

## Peak Smoothing Algorithm

Iterate through each data point in a given month.
Leave the first energy level where it is for now. Set the current max to be this first energy level.
If the second energy level is equal to or less than the current max, then leave it alone and continue.
If it is greater than the current max, then start from the beginning and redistribute the amount of energy greater than the current max to the preceding timepoints, while keeping each under the current max. Once done, if there still remains energy to be distributed, distribute among all preceding timepoints evenly and increase the current max accordingly. Repeat until the entire month has been smoothed.

We can also enforce a constraint that energy cannot be dissapated in advance more than 24 hours. This means that when we redistribute energy, we do not go to the beginning, we go only 24 hours back. This also means that the current max will need to be updated as we scroll our 24 hour window. According to our approach, it must be true that the current max is the first bar in the window.

```python
# Pass into this function a dataframe representing a single month
# Set window to a number, such as 24*4, to enforce it.
def smooth_peaks(dataframe, window=math.inf):
    dataframe = dataframe.copy()
    timepoints = np.asarray(dataframe['avg_power'])
    curr_max = timepoints[0]
    
    for t in range(len(timepoints)):
        if timepoints[t] > curr_max:
            surplus = timepoints[t] - curr_max
            timepoints[t] = curr_max
            
            # Redistribute to previous timepoints without increasing curr_max.
            for i in range(max(t-window, 0), t):
                if surplus > 0 and timepoints[i] < curr_max:
                    delta = min(curr_max-timepoints[i], surplus)
                    timepoints[i] += delta
                    surplus -= delta
            
            # If necessary, distribute remainder evenly and adjust curr_max.
            delta = surplus/(t+1 - max(t-window, 0))
            for i in range(max(t-window, 0), t+1):
                timepoints[i] += delta
            curr_max += delta
            
            # Adjust curr_max according to our window
            if window != math.inf:
                curr_max = min(curr_max, timepoints[max(t-window, 0)])
            
    return timepoints
```

```python
january = data.copy()
january['month'] = january['timestamp'].apply(lambda x: int(x.format('M')))
january = january[january['month']==1]

january['smoothed'] = smooth_peaks(january)
january['constrained_smooth'] = smooth_peaks(january, window=24*4)
january.to_csv("january.csv")
```

```python
plt.figure(figsize=(16, 12))
plt.title('January Energy Usage')
plt.plot(range(len(january['timestamp'])), january['avg_power'], label="Original")
plt.plot(range(len(january['timestamp'])), january['smoothed'], label="Smoothed")
plt.plot(range(len(january['timestamp'])), january['constrained_smooth'], label="Constrained Smoothed")
plt.legend()
plt.show()
```

## Smoothed demand charges (no constraint on when storage can begin)

```python
smoothed = pd.DataFrame()
for i in range(1, 13):
    monthdf = data.copy()
    monthdf['month'] = monthdf['timestamp'].apply(lambda x: int(x.format('M')))
    monthdf = monthdf[monthdf['month']==i]
    monthdf['smoothed'] = smooth_peaks(monthdf)
    monthdf['constrained'] = smooth_peaks(monthdf, window=24*4)
    smoothed = smoothed.append(monthdf)
```

```python
# Uncontrained smoothed demand charges.
compute_demand_charge(smoothed, 'smoothed')
```

```python
# 24Hr constrained smoothed demand charges.
compute_demand_charge(smoothed, 'constrained')
```

# Extrapolation to January 2018

Tutorial referenced: https://machinelearningmastery.com/time-series-forecasting-long-short-term-memory-network-python/

```python
# Set up inputs and outputs.

# Convert each timestamp to a single number representing the time.
X = np.asarray([data['timestamp']])
y = np.asarray([data['avg_power']])

# Set missing values to be 0.
y[np.isnan(y)] = 0
N = max(X.shape)

y = y.reshape([-1,1])

# Make our series stationary by removing diffs.
diffs = []
for i in range(7*24*4):
    diffs += [y[i]]
for i in range(7*24*4, N):
    diffs += [y[i] - y[i-7*24*4]]
    
y = np.asarray(diffs)

# Scale y to be in range [-1, 1].
scaler = MinMaxScaler(feature_range=(-1,1))
y = scaler.fit_transform(y)

# Roll y so that we can train on our time sequence.
y_shifted = np.roll(y, 7*24*4)
input_data = np.hstack([y_shifted, y])
```

```python
def fit(data, batch_size, epochs, neurons):
    X, y = data[:, 0:-1], data[:, -1]
    X = X.reshape(X.shape[0], 1, X.shape[1])
    model = Sequential()
    model.add(LSTM(neurons, batch_input_shape=(batch_size, X.shape[1], X.shape[2]), stateful=True))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')
    for i in range(epochs):
        print('Epoch #', i)
        model.fit(X, y, epochs=1, batch_size=batch_size, verbose=0, shuffle=False)
        model.reset_states()
    return model
```

```python
model = fit(input_data, 1, 50, 1)
```

```python
prediction = diffs[-7*24*4:] + [0]*31*24*4
for i in range(7*24*4, len(prediction)):
    prediction[i] = model.predict(prediction[i-7*24*4].reshape([1,1,1]), batch_size=1)

#prediction = scaler.inverse_transform(prediction)

prediction_recovered = []
for i in range(7*24*4):
    prediction_recovered += [prediction[i]]
for i in range(7*24*4, len(prediction)-7*24*4):
    prediction_recovered += [prediction_recovered[i-7*24*4]+prediction[i]]
    
prediction_recovered = np.asarray(prediction_recovered)
prediction_recovered[prediction_recovered<0] = 0
```

```python
plt.figure(figsize=(16, 12))
plt.plot(range(len(prediction_recovered)), prediction_recovered)
plt.title("Predicted Energy Usage for January 2018")
plt.show()
```

## Compute Forecasted Savings for January 2018

```python
prediction_recovered = np.asarray(prediction_recovered).reshape([-1,1])
jan18 = data['timestamp'][0:31*24*4]
jan18 = jan18.apply(lambda x: x.replace(year=2018))
jan18 = np.asarray(jan18).reshape([-1,1])

result = np.hstack([jan18, prediction_recovered])
result = pd.DataFrame()

result['index'] = range(len(jan18))
del result['index']

result['Time (Interval beginning)'] = jan18
result['timestamp'] = jan18
result['avg_power'] = prediction_recovered
result['Average Power (kW)'] = prediction_recovered
print('Original',compute_demand_charge(result, 'Average Power (kW)')['Demand Charge'][0])

result2 = result.copy()
result2['month'] = result2['timestamp'].apply(lambda x: int(x.format('M')))
result2 = result2[result2['month']==1]

result2['smoothed'] = smooth_peaks(result2)
result2['constrained_smooth'] = smooth_peaks(result2, window=24*4)

print('Smoothed',compute_demand_charge(result2, 'smoothed')['Demand Charge'][0])
print('Constrained Smooth',compute_demand_charge(result2, 'constrained_smooth')['Demand Charge'][0])

pd.DataFrame(result).to_csv("prediction.csv")
```
